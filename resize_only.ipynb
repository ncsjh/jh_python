```python
import os
import shutil
import numpy as np
from glob import glob
import tensorflow as tf
from PIL import Image, ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```python
cls=os.listdir('./FacialClassification/')
cls.remove('.ipynb_checkpoints')
```

```python
for cl in cls:
    path=f'./FacialClassification/{cl}/'
    print(f'경로 : {path}')
    print(f'{cl} 폴더 내 파일 불러오기')
    li=os.listdir(path)
    print(f'{cl} 폴더 내 파일 리사이즈 시작')
    w=len(li)
    c=0
    if cl=='분노':
        for file in li[46656:]:
            try:
                c+=1
                print('\r', int(c/w*100),'%', '  ', c,'/',w, file, end='')
                dest=path+file
                img=Image.open(dest)
                img=img.resize((600,600))
                img.save(dest)
            except:
                print(file, ' 파일에서 문제가 발생했습니다.')
                print('\r', int(c/w*100),'%', '  ', c,'/',w, file, end='')
    else : 
        for file in li:
            try:
                c+=1
                print('\r', int(c/w*100),'%', '  ', c,'/',w, file, end='')
                dest=path+file
                img=Image.open(dest)
                img=img.resize((600,600))
                img.save(dest)
            except:
                print(file, ' 파일에서 문제가 발생했습니다.')
                print('\r', int(c/w*100),'%', '  ', c,'/',w, file, end='')
```
